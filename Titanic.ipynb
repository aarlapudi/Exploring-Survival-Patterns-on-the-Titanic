```python
# You can create a Ticket_GroupSize feature instead, which is much more meaningful for survival analysis.
# Do passengers in groups have higher survival rates?
# This usually shows families/groups survived more than solo travelers.

ticket_counts = df['Ticket'].value_counts()
df['Ticket_GroupSize'] = df['Ticket'].map(ticket_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Pairplot 

```python
numeric_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Ticket_GroupSize']


sns.pairplot(df[numeric_cols + ['Survived']], 
             hue='Survived',        # Color by Survived
             diag_kind='hist',      # Histogram on diagonal
             palette='Set1')       # Color palette        
plt.suptitle("Pairplot of Numeric Features by Survival", y=1.02)
plt.show()
```

# Heatmap

```python
numeric_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Ticket_GroupSize']

# Compute correlation matrix
corr_matrix = df[numeric_cols + ['Survived']].corr()

# Plot heatmap
plt.figure(figsize=(10,6))  # Figure size
sns.heatmap(corr_matrix, 
            annot=True,        # Show correlation values on each cell
            cmap='coolwarm',   # Color map: red = positive, blue = negative
            fmt=".2f")         # Format correlation values to 2 decimal places
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

# Histogram Plots

```python
# Age distribution

plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
# Pclass distribution
plt.figure(figsize = (8,5))
sns.histplot(df['Pclass'], bins = 20, kde = True)
plt.title('Pclass distribution of passengers')
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=10, kde=True)
plt.title("Fare Distribution of Passengers")
plt.show()
```

```python
# Ticket group size distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Ticket_GroupSize'], bins=15, kde=False)
plt.title("Ticket Group Size Distribution")
plt.show()
```

# Box plot

```python
# Age by Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Fare by Survival
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
# Fare by Pclass
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

# Scatter plot

```python
# Fare vs Age, colored by survival
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=1)
plt.title("Fare vs Age by Survival")
plt.show()
```

```python
# Ticket Group Size vs Fare, colored by survival
plt.figure(figsize=(8,5))
sns.scatterplot(x='Ticket_GroupSize', y='Fare', hue='Survived', data=df, alpha=0.7)
plt.title("Ticket Group Size vs Fare by Survival")
plt.show()
```
